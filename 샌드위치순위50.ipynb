```python
from urllib.parse import urljoin
```

```python
rank = []
main_menu = []
cafe_name = []
url_add = []
s_url="http://www.chicagomag.com"
temp = soup.find_all('div','sammy')
len(temp)
for i in temp:
    rank.append(i.find(class_='sammyRank').get_text())
    
    tmp_string = i.find(class_='sammyListing').get_text()
    main_menu.append(re.split(('\n|\r\n'), tmp_string)[0])
    cafe_name.append(re.split(('\n|\r\n'), tmp_string)[1])
    url_add.append(urljoin(s_url, i.find('a')['href']))
```

```python
# conda install tqdm
# 작업 진행 내용을 확인하는 라이브러리

from tqdm.notebook import tqdm
```

```python
import time

item_total = tqdm(df.index)
price = []
address = []

for n in item_total:
    url = df['URL'][n]
    req=Request(url, headers=header)
    html = urlopen(req)
    soup = BeautifulSoup(html, 'html.parser')
    
    line_data = soup.find('p', 'addy').get_text()
    price.append(line_data.split()[0][:-1])
    address.append(' '.join(line_data.split()[1:-2]))
    time.sleep(0.05) # 새로운 페이지를 열 시간을 주기 위한 시간을 준다.
```

```python
# folium 지도를 이용하여 marker 찍기
# 주소가 Multiple인 것은 제외하고 위도,경도 가져오기
# folium 지도를 가져올 때 중앙 경도, 위도를 어떻게 가져올지
import googlemaps
import numpy as np
import folium
```

```python
gmaps_key="키값"
gmaps = googlemaps.Client(key=gmaps_key) # 클라이언트 객체 생성
```

```python
# tqdm 사용
lat=[]
lng=[]

eter_val = tqdm(df.index)
for n in eter_val:
    if df['Address'][n] != 'Multiple':
        target_name=df['Address'][n]+ ' Chicago'
        gmaps_out = gmaps.geocode(target_name)
        
        item_loc = gmaps_out[0].get('geometry')
        lat.append(item_loc['location']['lat'])
        lng.append(item_loc['location']['lng'])
        
    else:
        lat.append(np.nan)
        lng.append(np.nan)
```

```python
station_address = []
station_lat = [] # 경도
station_lng = [] # 위도

for name in df['Address']:
    if name != 'Multiple':
        item = gmaps.geocode(name+'Chicago', language='eng')
        station_address.append(item[0].get('formatted_address'))
    
        item_loc = item[0].get('geometry')
        station_lat.append(item_loc['location']['lat'])
        station_lng.append(item_loc['location']['lng'])
    
    else:
        station_address.append(np.nan)
        station_lat.append(np.nan)
        station_lng.append(np.nan)
```

```python
df['Lat']=station_lat
df['Lng']=station_lng
```
